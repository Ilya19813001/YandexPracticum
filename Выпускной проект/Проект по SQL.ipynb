#### Задание 1. Посчитайте, сколько книг вышло после 1 января 2000 года.

```python
query = ''' 
    SELECT COUNT(DISTINCT book_id) AS cnt_books
    FROM books
    WHERE publication_date > '2000-01-01';
    ''' 

pd.io.sql.read_sql(query, con = engine)
```

После 1 января 2000 года вышло 819 книг. Пока нам не с чем сравнить . Знаем что сервис работает выпускает книги.Бизнес жив. 

#### Задание 2. Для каждой книги посчитайте количество обзоров и среднюю оценку.

```python
query = '''

SELECT b.book_id AS id,
       b.title AS bt,
       COUNT(DISTINCT r.username) AS cnt_rw,
       ROUND(AVG(rt.rating),2) AS AVG_rtn
FROM books AS b
FULL OUTER JOIN reviews AS r ON b.book_id = r.book_id
FULL OUTER JOIN ratings AS rt ON b.book_id = rt.book_id
GROUP BY b.title, b.book_id
ORDER BY cnt_rw DESC;
 ''' 

pd.io.sql.read_sql(query, con = engine)
```

Мы получили список всех книг, у которых есть рейтинги и обзоры, и у которых нет ревью и рейтинга. Сумерки  7 обзоров и  средний рейтинг 3,66.

#### Задание 3. Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры.

```python
query = '''

SELECT p.publisher AS pub,
       COUNT(DISTINCT b.book_id) AS max_cnt_b
       
FROM publishers AS p
JOIN books AS b ON p.publisher_id = b.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY max_cnt_b DESC
LIMIT 1;
 ''' 

pd.io.sql.read_sql(query, con = engine)
```

Издательство Penguin Books выпустило наибольшее число книнг толще 50 страниц - 42 книги

#### Задание 4. Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками.

```python
query = '''

SELECT a.author AS ath,
       AVG(r.rating) AS avg_r
       
FROM authors AS a
JOIN books AS b ON a.author_id = b.author_id
JOIN ratings AS r ON b.book_id = r.book_id
GROUP BY a.author
HAVING COUNT(r.rating) > 50
ORDER BY avg_r DESC
LIMIT 1;
 ''' 

pd.io.sql.read_sql(query, con = engine)
```

Джоан Кэ́тлин Роулинг автор которому присвоен самы высокий средний балл 4,288462 . Получила мировое признание после выпуска серии романов о Гарри Поттере.

#### Задание 5. Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''

SELECT AVG(rw.rw_cnt) AS avg_cnt_rw       
FROM (
      SELECT username,
      COUNT(review_id) AS rw_cnt
      FROM reviews
      GROUP BY username
      HAVING username IN (
                          SELECT username
                          FROM ratings
                          GROUP BY username
                          HAVING COUNT(rating_id) > 50   
                        ) 
                 ) AS rw;
'''
pd.io.sql.read_sql(query, con = engine)
```

В среднем 24,33 обзора публикуют пользователи которые поставил больше 50 оценок. Активные пользователи  - почти 50% обзоров от количества оценок.
