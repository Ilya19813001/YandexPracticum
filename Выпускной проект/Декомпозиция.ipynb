# Анализ оттока клиентов  банка «Метанпром»

Мы работаем аналитиками данных в банке "Метанпром". К нам обратился product-менеджер с целью получения рекоменданций по сегменту клиентов, которые склонны уходить из банка.

## Цель исследования

    
Минимизировать отток клиентов из нашего банка с целью удержания текущей клиентской базы и ее расширения. Для этого проанализируем данные клиентов регионального банка и выделим портрет клиентов, которые склонны уходить из банка и которые склонны остаться. Предоставить рекомендации product-менеджеру для принятия им, соответствующих решений по данному вопросу.

**Задачи:**

•	Проведем исследовательский анализ данных,

•	Выделим портреты клиентов, которые склонны уходить из банка,

•	Выделим портреты клиентов, которые не склонны уходить из банка,

•	Сформулируем и проверим статистические гипотезы.



## Ход исследования

 - Загрузка данных 
 
 - Изучение общей информации и предобработка данных
 
   - изучим название колонок
   
   - наличие пропусков
   
   - наличие дубликатов
   
   - изучим типы данных
   
   - промежуточные выводы
 
 - Исследовательский анализ данных 
 
   - Анализ описательных статистик показателей
   
   - Анализ распределения признаков
   
   - Корреляционный анализ
   
   - Промежуточные выводы
 
 - Выведение портрета клиентов, которые склонны и не склонны уходить из банка
 
   - Анализ клиентов в разрезе доходов,  которые склонны и не склонны уходить из банка
   
   - Анализ клиентов в разрезе возрастов,  которые склонны и не склонны уходить из банка
 
 - Проверка гипотез
 
    Гипотеза о различии дохода между теми клиентами, которые ушли и теми, которые остались
   
    Гипотеза о различии возраста между теми клиентами, которые ушли и теми, которые остались.
    
 - Выводы и рекомендации.
